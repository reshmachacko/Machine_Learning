### Multivariate Analysis

#### LINE CHART

```python
data.plot.line(x='Sales_in_thousands',y='Price_in_thousands',cmap='spring');
```

```python
data.sort_values(by='Sales_in_thousands').plot.line(x='Sales_in_thousands',y='Price_in_thousands',cmap='spring');
```

```python
data.sort_values(by='year_resale_value').plot.line(x='year_resale_value',y='Price_in_thousands',cmap='ocean');
```

#### BAR PLOTS

```python
data.Manufacturer.value_counts().plot.bar(figsize=(6,3),cmap='PRGn')
```

```python
data.sort_values(by='Price_in_thousands')[:20].plot.bar('Manufacturer','Price_in_thousands',figsize=(6,2));
```

```python
data.sort_values(by='Power_perf_factor').plot.bar('Model','Power_perf_factor',figsize=(18,2),cmap='cool');
```

### AREA CHART

```python
data.plot.area('Manufacturer','Price_in_thousands',figsize=(6,4),cmap='Accent')
```

#### HORIZONTAL BAR CHART

```python
data.sort_values(by='Price_in_thousands')[:25].plot.barh('Model','Power_perf_factor',figsize=(6,11),cmap='inferno');
```

```python
data.sort_values(by='Horsepower')[:5].plot.barh('Manufacturer','Price_in_thousands',figsize=(3,4),cmap='summer');
```

#### SCATTER PLOT

```python
data.sort_values(by='Fuel_capacity').plot.scatter('Fuel_capacity','Fuel_efficiency',figsize=(3,3),cmap='Paired')
```
