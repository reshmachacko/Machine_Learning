#### loading dependencies

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

#### loading dataset

```python
data=pd.read_csv('./titanic (1).csv')
data.head()
```

#### Droping multiple columns

```python
data.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
data.head(10)
```

### GROUPBY

#### gendervise survival

```python
Sex_survived_sum=data.groupby(['Sex'])['Survived'].sum()
```

```python
Sex_survived_sum
```

```python
Sex_survived_avg=data.groupby(['Sex'])['Survived'].mean()*100
Sex_survived_avg
```

#### classvise Survival

```python
Pclass_survived_avg=data.groupby(['Pclass'])['Survived'].mean()*100
Pclass_survived_avg
```
