```python
data.Medal.unique()
```

### Lets create Seperate columns using if else statements for number of medals

```python
data['Gold']=[1 if medal=='Gold' else 0 for medal in data['Medal']]
data['Silver']=[1 if medal=='Silver' else 0 for medal in data['Medal']]
data['Bronze']=[1 if medal=='Bronze' else 0 for medal in data['Medal']]
```

```python
data
```

### 1. Which country has won the most number of medals

```python
country_vise_medal=data.groupby(['Team'])['Gold','Silver','Bronze'].sum()
country_vise_medal
```

```python
country_vise_medal['Total']=country_vise_medal['Gold']+country_vise_medal['Silver']+country_vise_medal['Bronze']
country_vise_medal
```

```python
country_vise_medal.sort_values(by='Total',ascending=False)
```

```python
country_vise_medal.nlargest(1, 'Total')
```

##### United States won the most number of medals

### 2. Country with no gold medal at all

```python
no_gold_medal_at_all=country_vise_medal[country_vise_medal['Gold']==0]
no_gold_medal_at_all
```

```python
list_no_gold_medal_at_all=list(no_gold_medal_at_all.index)
list_no_gold_medal_at_all
```

```python
print('Number of countries with no gold medal at all:',len(list_no_gold_medal_at_all))
```
